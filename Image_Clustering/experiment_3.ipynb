```python
from keras.datasets import mnist
import numpy as np
np.random.seed(10)

from time import time
import numpy as np
import keras.backend as K
from keras.engine.topology import Layer, InputSpec

from keras.models import Model
from keras.optimizers import SGD
from keras import callbacks
from keras.initializers import VarianceScaling
from sklearn.cluster import KMeans
from sklearn import metrics
from keras import layers
from keras.layers import Input, Dense, Conv2D, MaxPooling2D, UpSampling2D, Flatten, Reshape, Conv2DTranspose

from encoder_function import autoencoder, autoencoderConv2D_1, autoencoderConv2D_2, ClusteringLayer
from clustering_model_function import Basic_Kmeans, AutoEncoder_Kmeans, DEC_Kmeans, Conv_AutoEncoder_Kmeans,Conv_DEC_Kmeans,Conv_b_DEC_Kmeans
```

```python
import cv2
from keras.preprocessing.image import img_to_array
import os
root_path = './label_6/'
images = []
labels = []
for folder in os.listdir(root_path):
    if 'DS_Store' in folder:
        continue
    for f_name in os.listdir(os.path.join(root_path, folder)):
        if 'DS_Store' in f_name:
            continue
        image = cv2.imread(os.path.join(root_path, folder, f_name))
        image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
        image = cv2.resize(image, (28,28))
        image = image.astype("float") / 255.0
        image = img_to_array(image)
        images.append(image)
        labels.append(folder)

images = np.array(images)
x = images.reshape((images.shape[0], -1))
y = labels
```

```python
model1 = Basic_Kmeans(images, labels=y, n_init=12, n_jobs=-1)
```

```python
model2 = AutoEncoder_Kmeans(images, labels=y, n_init=12, n_jobs=-1,
                            save_dir='./results/screen_clustering/',
                            learning_rate=0.01,
                            momentum=0.9,
                            pretrain_epochs=100,
                            batch_size=200)
```

```python
model3 = DEC_Kmeans(images,labels, n_init=12, n_jobs=-1,
            save_dir='./results/screen_clustering/',
            learning_rate=0.01,
            momentum=0.9,
            pretrain_epochs=100,
            batch_size=200,
            maxiter=1000,
            update_interval=200,
            tol=0.001)
```
